# MatplotLib

Matplotlib is a ploting Library for the python programing language and its numerical mathematics extention Numpy. It provides object oriented API for embedding plots into application using general purpose GUi toolkits like Tkinter, wcpython,Qt or GTK+


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
a = np.arange(10,100,5)
b = np.arange(100, 190, 5)
```

```python
plt.scatter(a,b)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("XY Plot")
```

```python
b = a*a
```

```python
plt.plot(a,b,'r--*')
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title('2D plot')
```

```python
plt.subplot(2,2,1)
plt.plot(a,b,"r--")
plt.subplot(2,2,2)
plt.plot(a,b,"g*-")
plt.subplot(2,2,3)
plt.plot(a,b,"bo")
```

```python
x =np.arange(1,7*np.pi,.1)
y = np.sin(x)
plt.plot(x,y)

```

```python
x = np.arange(1,5*np.pi, .1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y1,'r--')
plt.subplot(2,1,2)
plt.plot(x,y2,'b')
```

```python
x =[6,18,15]
y =[4,15,5]
x1 = [3,11,9]
y2 = [5,10,18]
```

```python
plt.bar(x,y,color = 'g')
plt.bar(x1,y2)
```

```python
#Histogram
v= np.array([23,5,6,8,13,42,34,21,16,14,32])
```

```python
#vary bins for exact graph
plt.hist(v,bins = 5)
```

# Box Plot using Matplot

```python
data= [np.random.normal(0,std,100) for std in range(1,4)]
data
```

```python
plt.boxplot(data,vert= True,patch_artist=True)
```

```python
lable = ["python", "C++", "Java", "Ruby"]
size = [215, 355, 155,352]
color = ["yellow","green","blue","red"]
explode = [.2,0,0,0]
plt.pie(size,explode=explode,labels = lable,colors=color,autopct='%1.4f%%',shadow=True)
```
