# Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

# heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# jointplot

```python
sns.jointplot(x ="tip",y ="total_bill",data=df,kind = 'hex')
```

```python
sns.jointplot(x ="tip",y ="total_bill",data=df,kind = 'reg')
```

# pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue = 'smoker')
```

# dist plot

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
ir = sns.load_dataset('iris')
```

```python
ir.head()
```

```python
sns.heatmap(ir.corr())
```

```python
sns.jointplot(x ="petal_length", y ="sepal_width",data =ir,kind = "hex")
```

```python
sns.jointplot(x ="sepal_length", y ="sepal_width",data =ir,kind = "reg")
```

```python
sns.pairplot(ir, hue="species")
```

# countplot

```python
sns.countplot(x = 'species', data = ir)
```

```python
sns.barplot(x = 'species', y = 'sepal_length', data = ir)
```

```python
sns.barplot(x = 'species', y = 'petal_length', data = ir)
```

```python
sns.boxplot(x = 'species', y = 'petal_length', data = ir)
```

```python
sns.barplot(x = 'species', y = 'petal_length', data = ir,hue="sepal_width")
```

# violinplot

```python
sns.violinplot(x = 'species', y = 'petal_length', data = ir)
```
