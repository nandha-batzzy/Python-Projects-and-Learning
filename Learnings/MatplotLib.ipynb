# MatplotLib

Matplotlib is a ploting Library for the python programing language and its numerical mathematics extention Numpy. It provides object oriented API for embedding plots into application using general purpose GUi toolkits like Tkinter, wcpython,Qt or GTK+


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
a = np.arange(10,100,5)
b = np.arange(100, 190, 5)
```

```python
plt.scatter(a,b)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("XY Plot")
```

```python
b = a*a
```

```python
plt.plot(a,b,'r--*')
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title('2D plot')
```

```python
plt.subplot(2,2,1)
plt.plot(a,b,"r--")
plt.subplot(2,2,2)
plt.plot(a,b,"g*-")
plt.subplot(2,2,3)
plt.plot(a,b,"bo")
```

```python
x =np.arange(1,7*np.pi,.1)
y = np.sin(x)
plt.plot(x,y)

```

```python
x = np.arange(1,5*np.pi, .1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y1,'r')
plt.subplot(2,1,2)
plt.plot(x,y2,'b')
```
